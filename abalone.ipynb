BACKGROUND OF THE DATA : 

According to Wikipedia, "Abalone" is a common name for any member of the family Haliotidae, a group of small to very large sea snails, also known as marine gastropod mollusks. The current dataset is an updated version of a study conducted by a scientific team in 1994. The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings under a microscope. Other, easier-to-obtain physical measurements can also be used to estimate its age.

```python
#Dummies
ab1 = pd.get_dummies(ab)
ab1.head()
```

```python
y=ab1['Rings']
ab2=ab1.drop(["Rings"],axis=1)
x=ab2

# getting the shapes
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# getting the shapes
print("Shape of x_train :", x_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of y_test :", y_test.shape)
```

```python

# MODELLING
# RANDOM FOREST REGRESSOR

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

# evaluation
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE :", rmse)

# r2 score
r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```
